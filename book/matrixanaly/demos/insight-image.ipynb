# Images

```julia
using FundamentalsNumericalComputation
```

The `Images` package has many functions for image manipulation, and `TestImages` has some standard images to play with.

```julia
im = testimage("mandrill")
```

The variable `im` is a matrix...

```julia
size(im)
```

...but its entries are colors, not numbers.

```julia
im[100,10]
```

```julia
eltype(im)
```

It's possible to extract matrices of red, green, and blue intensities, scaled from 0 to 1.

```julia
R,G,B = red.(im),green.(im),blue.(im)
extrema(B)
```

Instead, here we convert the entries to gray, each pixel again scaled from 0 to 1.

```julia
Gray.(im)
```

In order to do our usual operations, we need to tell Julia that we want to interpret the elements of the image matrix as floating-point values.

```julia
A = Float64.(Gray.(im))
```

We can convert any such matrix to grayscale again in order to view it as an image.

```julia
Gray.(reverse(A,dims=1))
```
